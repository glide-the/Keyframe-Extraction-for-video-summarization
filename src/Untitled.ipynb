```python
import sys
sys.path.append('/media/gpt4-pdf-chatbot-langchain/Keyframe-Extraction-for-video-summarization/src')
import pickle
from tqdm import tqdm  # Import the tqdm function or class
import numpy as np
from extraction.init_center import kmeans_init
import matplotlib.pyplot as plt  # 导入matplotlib库

# Get lens segmentation data
scenes_path='/home/dmeck/Downloads/兔兔没有牙.mp4.scenes.txt'
features_path='/home/dmeck/Downloads/兔兔没有牙-features/features.pkl'
```

```python

number_list = []
with open(scenes_path, 'r') as file:
    lines = file.readlines()
    for line in lines:
        print(line)
        numbers = line.strip().split(' ')
        print(numbers)
        number_list.extend([int(number) for number in numbers])

# Read inference data from local
with open(features_path, 'rb') as file:
    features = pickle.load(file)

features = np.asarray(features)
# Clustering at each shot to obtain keyframe sequence numbers
keyframe_index = []
for i in range(0, len(number_list) - 1, 2):
    start = number_list[i]
    end = number_list[i + 1] 
    sub_features = features[start:end] 
     # Results of the selection of the initial center
    clusters, centers = kmeans_init(sub_features)

    # Assuming clusters and centers have been calculated in your loop
    # clusters.shape should be (119,) and centers.shape should be (10, 768)
    
    # Plot the number of data points in each cluster
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.hist(clusters, bins=range(11), alpha=0.5, color='b', rwidth=0.8)
    plt.xlabel('Cluster Label')
    plt.ylabel('Number of Data Points')
    plt.title('Distribution of Data Points in Clusters')
    
    # Plot the number of centers in each cluster
    plt.subplot(1, 2, 2)
    num_centers_in_clusters = [len(set(clusters[clusters == i])) for i in range(10)]
    plt.bar(range(10), num_centers_in_clusters, alpha=0.5, color='g')
    plt.xlabel('Cluster Label')
    plt.ylabel('Number of Centers')
    plt.title('Number of Centers in Each Cluster')
    
    plt.tight_layout()
    
    # Plot the shape of the centers
    plt.figure(figsize=(8, 4))
    plt.plot(range(768), centers[0], label='Cluster 0', linestyle='-', marker='o', markersize=3)
    plt.plot(range(768), centers[1], label='Cluster 1', linestyle='-', marker='o', markersize=3)
    plt.xlabel('Feature Index')
    plt.ylabel('Center Value')
    plt.title('Centers of Clusters')
    plt.legend()
    
    plt.show()
     
```
